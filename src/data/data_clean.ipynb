```python
import pandas as pd
```

```python
df = pd.read_excel("data.xlsx",sheet_name="GSI 2023 summary data")
df
```

REMOVING DUPLICATES

```python
df.drop_duplicates()
```

Removing extra rows on top

```python
# Drop completely empty columns
df = df.dropna(axis=1, how='all')

# Remove the first two rows
df = df.iloc[1:].reset_index(drop=True)

# Set the new header row (previous row 3, now row 1)
df.columns = df.iloc[0]  # Assign new column names
df = df[1:].reset_index(drop=True)  # Remove the row used as headers

df
```

```python
df = df[['Country',
         'Population', 
         'Region',
         'Estimated prevalence of modern slavery per 1,000 population',
         'Estimated number of people in modern slavery',
         'Governance issues',
         'Effects of conflict',
         'Total Vulnerability score (%)',
         'Risk factors, such as attitudes, social systems, and institutions that enable modern slavery are addressed (%)',
         'Government response total (%)',
        "Survivors of slavery are identified and supported to exit and remain out of modern slavery (%)"    # Victim support metric
         ]]
df
```

```python
for x in df.index:
    if df.loc[x,'Total Vulnerability score (%)'] == 'NaN'  or df.loc[x,'Government response total (%)'] == 'N/A':
        df.drop(x, inplace=True)
df.reset_index()
df
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
